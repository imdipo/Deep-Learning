```python
# drop data null
data = data.dropna(how='any',axis=0)
data
```

```python
siswa = data.drop(['Unnamed: 0', 'EthnicGroup', 'ParentEduc', 'NrSiblings', 'TransportMeans', 'ParentMaritalStatus', 'PracticeSport'], axis=1)
```

```python
# relationship analysis
data['WklyStudyHours'].unique()
```

```python
# Mengubah nilai string menjadi numerik untuk kolom biner
siswa['Gender'] = siswa['Gender'].map({'female':0, 'male':1})
siswa['LunchType'] = siswa['LunchType'].map({'standard':0, 'free/reduced':1})
siswa['TestPrep'] = siswa['TestPrep'].map({'none':0, 'completed':1})
siswa['IsFirstChild'] = siswa['IsFirstChild'].map({'no':0, 'yes':1})

# Menggunakan One-Hot Encoding untuk kolom 'WklyStudyHours'
siswa_encoded = pd.get_dummies(siswa, columns=['WklyStudyHours'])

# Menghapus kolom asli 'WklyStudyHours' dari DataFrame siswa
siswa = siswa.drop(columns=['Gender'])
siswa = siswa.drop(columns=['LunchType'])
siswa = siswa.drop(columns=['TestPrep'])
siswa = siswa.drop(columns=['IsFirstChild'])
siswa = siswa.drop(columns=['WklyStudyHours'])
siswa = siswa.drop(columns=['MathScore'])
siswa = siswa.drop(columns=['ReadingScore'])
siswa = siswa.drop(columns=['WritingScore'])


# Menggabungkan DataFrame asli dengan DataFrame yang sudah di-encode
siswa_combined = pd.concat([siswa, siswa_encoded], axis=1)
```

```python
# membuat korelasi
correlation = siswa_combined.corr()
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns, annot=True)
plt.show()
```

```python
# Dari heatmap diatas kita dapat memahami beberapa hal
# 1. MathScore memiliki hubungan positif dengan ReadingScore sebanyak 0,82 dan 0,81 pada WritingScore
# 2. ReadingScore memiliki hubungan positif dengan MathScore sebanyak 0,95 
# 3. mayoritas data memiliki hubungan negatif
```

```python
fig, axarr = plt.subplots(2, 2, figsize=(20, 20))
data.groupby('IsFirstChild')['MathScore'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][0], fontsize=12)
axarr[1][0].set_title("Anak_Pertama Vs Math", fontsize=18)
data.groupby('TestPrep')['MathScore'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][1], fontsize=12)
axarr[1][1].set_title("Persiapan_Ujian Vs Math", fontsize=18)
data.groupby('Gender')['MathScore'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][1], fontsize=12)
axarr[1][1].set_title("Jenis_Kelamin Vs Math", fontsize=18)
data.groupby('LunchType')['MathScore'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][0], fontsize=12)
axarr[1][1].set_title("Makan_siang Vs Math", fontsize=18)

plt.subplots_adjust(hspace=1.0)
plt.subplots_adjust(wspace=0.5)
sns.despine()
```
